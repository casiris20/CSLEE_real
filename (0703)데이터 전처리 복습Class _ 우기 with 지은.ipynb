### 6. debt에 0부터 값이 증가하도록 채우기

```python
count = len(df)   #1338
```

```python
A = [] # 빈 리스트를 만들어라..

# 빈 리스트에 넣어야 하는 규칙을 만들어라.
for i in range(count):
    
    A.append(i)

A
```

```python
A = np.array([A]) # 각 행들끼리 구분이 안되는 1D 차원이다.
A.shape # array는 transpose가 안됀다. 왜냐하면 '행렬'개념이 아니기 때문이다.
```

```python
A = np.matrix(A)  # 행렬 기준으로 바꾸려면은 matrix를 써야한다.
A

# ex) [1,2] 는 1X2 행렬이다. 따라서 [1,2,3,4,5,,,,,,,,,,1338]은 1X1338 행렬인 것이다.
# array(1,2) = [1,2]      는 (2,)이 나오는데, 데이터가 2개가 있다는 말이다.. (1과 2... so, 2개의 데이터)
# array([1,2]) = [1,2]    는 (1,2)이 나오는데 [1,2]가 1개이고 그안에 데이터가 2개가 들어갔다. 즉 1X2 행렬이라는 말이다.  이는 matrix(1,2)과 같다.

# matrix(1,2) = [1,2]  라는 뜻! 1X2 행렬이라는 말

# matrix([[   0,    1,    2, ..., 1335, 1336, 1337]])     ===> 1X1338 행렬이다.. 
```

```python
#a=np.array([[1,2],[2,3]])
#a.shape
a=np.array([[1,2,3],[2,3,3]])
a.shape     ### =====> 2개의 데이터가 있고, 각각 3개의 데이터를 가지고 있다.
```

```python
A = np.transpose(A)  # 1X1338 행렬을 1338X1 행렬로 변환하라
A.shape # matrix는 '행렬'개념이므로, transpose가 가능하다.
```

```python
debt = A
df['debt'] = debt
df.head()
```
