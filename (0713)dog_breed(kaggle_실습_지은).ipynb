```python
import os
import tensorflow as tf
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from keras.preprocessing.image import img_to_array,load_img,ImageDataGenerator
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from keras.callbacks import ModelCheckpoint,EarlyStopping
from tensorflow.keras.applications import VGG16

np.random.seed(3)
tf.compat.v1.set_random_seed(3)
```

```python
DATA_DIR = './dog_breed'


TRAIN_DIR = DATA_DIR + '/train'                           
TEST_DIR = DATA_DIR + '/test'                             

TRAIN_CSV = DATA_DIR + '/labels.csv'                     
#TEST_CSV = DATA_DIR + '/submission.csv' 
```

```python
data = pd.read_csv(TRAIN_CSV)
data.head(10)
```

```python
labels_names=data["breed"].unique() #unique() : 유일한 값 찾기
labels_sorted=labels_names.sort()

# 이 함수는 지정된 파일 안의 모든 단어로 구성된 어휘집을 만든다.
labels = dict(zip(range(len(labels_names)),labels_names))
labels 
```

```python
num_images = len(data["id"])
print('Train data의 이미지 개수:', num_images)
no_labels=len(labels_names)
print('Train data의 견종 개수:', no_labels)
```

```python
data.breed.value_counts()
```

```python
path_img=[]

for i in range(len(data["id"])):
    temp = TRAIN_DIR + "/" + str(data.id[i])+".jpg"
    path_img.append(temp)
    
data['path_img'] = path_img
data.head()
```

```python
###ASK!!!
fig, axes = plt.subplots(nrows=4, ncols=5, figsize=(15, 15),
                        subplot_kw={'xticks': [], 'yticks': []})

for i, ax in enumerate(axes.flat):
    ax.imshow(plt.imread(data.path_img[i]))
    ax.set_title(data.breed[i])
plt.tight_layout()
plt.show()
```

```python
top_20=list(data.breed.value_counts()[0:20].index)
top_20
```

```python
df=data[data.breed.isin(top_20)]
df.shape
```

```python
df.index = [i for i in range(len(df))]
df.head()
```

```python
from PIL import Image 
#Python Imaging Library(PIL)은 파이썬 인터프리터에
#다양한 이미지 파일 형식을 지원하고 강력한 이미지 처리와 
#그래픽 기능을 제공하는 자유-오픈 소스 소프트웨어 라이브러리이다.
```

```python
im_list = [Image.open(df.path_img[i]).size for i in range(len(df))]

# 이미지의 크기가 제각각이다.

print(min(im_list))
print(max(im_list))
print(np.mean(im_list))
```

```python
img_label=df.breed
img_label=pd.get_dummies(df.breed)
#get_dummies해주니 결측값을 제외한 값들이 0과 1로 변환되었다. => 원핫인코딩 (백터화)
img_label
```

```python
img_pixel=np.array([img_to_array(load_img(img, target_size=(300, 300))) 
                    for img in df.path_img.values.tolist()])
#Python list 객체가 필요할 때 tolist() 메서드를 사용하여 
#배열을 목록으로 변환 할 수 있다
```

```python
X=img_pixel
y=img_label.values
print(X.shape)
print(y.shape)
```

```python
X_train,X_val,y_train,y_val=train_test_split(X,y,test_size=0.4)
print(X_train.shape)
print(y_train.shape)
print(X_val.shape)
print(y_val.shape)
```

```python
X_train = X_train / 255
X_val = X_val / 255
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), 
                 input_shape=(300, 300, 3), activation='relu'))
model.add(Conv2D(32, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=2, strides=2))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(64,  activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(20, activation='softmax')) # 20개의 품종을 구분하는 것이므로 20!
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.summary()
```
