```python
from bs4 import BeautifulSoup
import requests # url을 request 한다.
import pandas as pd
```

```python
# 크롤링할 웹 사이트 정보
base_url = 'https://www.tripadvisor.co.kr' # 크롤링할 사이트 ( 기본  url)
seoul_url = '/Attractions-g294197-Activities-c26-t142-Seoul.html'

# get html
response = requests.get(base_url+seoul_url) # 이 주소의 '문서'를 줘! 라고 요청을 한다.
response  # response 200 : 성공했다는 뜻
```

```python
soup = BeautifulSoup(response.text, 'html.parser') # 태그별로 줄이 바뀌어져 있고, 정돈된 형태로 parsing이 된다.
```

```python
name_lists = soup.findAll('h2')#('div', {'class' : '_6sUF3jUd'})

rest_name_list = []

for item in name_lists:
    temp = str(item)
    name = item.text  # item에 있는 텍스트만 가져와!!! (item.text)   # 광고는 제외해!!! (.split('.'))
#     print(name)
    rest_name_list.append(name)
    
rest_name_list
```

```python
##########################################################3
url_lists = soup.findAll('div', {'class' : '_6sUF3jUd'})
rest_url_list = []


for item in url_lists:
    # 링크 가져오기
    hrefs = item.find('a')
    href = hrefs.get('href')
    full_url = base_url + href
    rest_url_list.append(full_url)
    #print(hrefs)
    #print(href)
    #print(full_url)
```

```python
result = pd.DataFrame(data={'[NAME]' : rest_name_list, '[URL]' : rest_url_list})
```
