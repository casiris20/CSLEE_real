## 시간 나눈거

```python
data.dtypes
data['datetime'] = data['datetime'].astype('datetime64')
data.dtypes
```

```python
data['year'] = data['datetime'].dt.year
data['month'] = data['datetime'].dt.month
data['day'] = data['datetime'].dt.day
data['hour'] = data['datetime'].dt.hour
data['minute'] = data['datetime'].dt.minute
data['second'] = data['datetime'].dt.second
```

# 시간별 유형별 사용자수 (회원 : 출근시간에 많음, 비회원 : 오후에 많음)

```python
sns.lineplot(data=data, x=data['hour'], y="count")
sns.lineplot(data=data, x=data['hour'], y="casual")
sns.lineplot(data=data, x=data['hour'], y="registered")
sns.label()
```

# 계절별 사용자수, 따듯할때 사용자가 많음

```python
sns.barplot(x = 'season', y = 'count', data = data, estimator = np.average, palette='rocket')
plt.ylabel('Average Count')
plt.show()
```

# 월별 사용자수 (여름에 너무 더우면 사용자 없음)

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(data=data, x=data["month"], y=data["atemp"], size=data["count"], sizes=(0, 100))
```

# 날씨별 사용자수 (날씨가 안좋을수록 사용자가 적다.)

```python
sns.relplot(x="hour", y="count", hue="weather", data=data);
```
